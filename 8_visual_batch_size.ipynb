```python
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots
# plt.style.use(['science','grid','vibrant'])


# 读取CSV文件
data = pd.read_csv('tmp_files/tem.csv',sep='\t')

with plt.style.context(['science','grid','vibrant']):
    plt.figure()
    plt.plot(data['Batch Size'], data['MSLS-val'], label='MSLS-val', 
                marker='o',  linestyle='-.', )
    plt.plot(data['Batch Size'], data['Pitts30k'], label='Pitts30k', 
                marker='s', linestyle='-.', )
    plt.plot(data['Batch Size'], data['Tokyo24/7'], label='Tokyo24/7', 
                marker='^', linestyle='-.', )
    plt.plot(data['Batch Size'], data['SPED'], label='SPED',
                marker='D', linestyle='-.', )

    # 设置x轴刻度为实际的batch size值
    plt.xticks(data['Batch Size'], data['Batch Size'])

    plt.xlabel('Batch Size')
    plt.ylabel('Performance (R1)')
    plt.legend(frameon=False)

    plt.tight_layout()
    plt.savefig('images/batch_size_acc.png', dpi=600)
    plt.show()
    
```
