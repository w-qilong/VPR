```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scienceplots
import matplotlib

print(matplotlib.get_configdir())

df=pd.read_excel('tmp_files/epoch_effect.xlsx')

# 
with plt.style.context(['science', 'grid','vibrant']):

    # xEpoch
    epochs = df['Epoch'].values

    # 14figsize
    fig, axes = plt.subplots(1, 4, figsize=(12, 3))

    # xEpoch
    epochs = df['Epoch'].values

    # MSLS
    axes[0].plot(epochs, df['MSLS-val R@1'], 'o-', label='R@1')
    axes[0].plot(epochs, df['MSLS-val R@5'], 's-', label='R@5')
    axes[0].plot(epochs, df['MSLS-val R@10'], '^-', label='R@10')
    axes[0].set_title('MSLS-val')
    axes[0].legend(frameon=False)
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Recall (\\%)')
    axes[0].set_xticks(epochs)

    # Nordland
    axes[1].plot(epochs, df['Nordland R@1'], 'o-', label='R@1')
    axes[1].plot(epochs, df['Nordland R@5'], 's-', label='R@5')
    axes[1].plot(epochs, df['Nordland R@10'], '^-', label='R@10')
    axes[1].set_title('Nordland')
    axes[1].legend(frameon=False)
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Recall (\\%)')
    axes[1].set_xticks(epochs)

    # Pitts
    axes[2].plot(epochs, df['SPED R@1'], 'o-', label='R@1')
    axes[2].plot(epochs, df['SPED R@5'], 's-', label='R@5')
    axes[2].plot(epochs, df['SPED R@10'], '^-', label='R@10')
    axes[2].set_title('SPED')
    axes[2].legend(frameon=False)
    axes[2].set_xlabel('Epoch')
    axes[2].set_ylabel('Recall (\\%)')
    axes[2].set_xticks(epochs)

    # Tokyo24/7
    axes[3].plot(epochs, df['Tokyo24/7 R@1'], 'o-', label='R@1')
    axes[3].plot(epochs, df['Tokyo24/7 R@5'], 's-', label='R@5')
    axes[3].plot(epochs, df['Tokyo24/7 R@10'], '^-', label='R@10')
    axes[3].set_title('Tokyo24/7')
    axes[3].legend(frameon=False)
    axes[3].set_xlabel('Epoch')
    axes[3].set_ylabel('Recall (\\%)')
    axes[3].set_xticks(epochs)

    # 
    plt.tight_layout()
    plt.savefig('images/start_epoch.png',dpi=600)
    # 
    plt.show()
```
