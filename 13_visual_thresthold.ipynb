```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scienceplots
import matplotlib

print(matplotlib.get_configdir())

df=pd.read_excel('tmp_files/threshold_effect.xlsx')
print(df.head())

# 
with plt.style.context(['science', 'grid','vibrant']):

    # xEpoch
    hbm_sizes = [str(size) for size in df['thresholds'].values]

    # 14figsize
    fig, axes = plt.subplots(1, 4, figsize=(12, 3))

    # MSLS
    axes[0].plot(hbm_sizes, df['Tokyo24/7 R@1'], 'o-', label='R@1')
    axes[0].plot(hbm_sizes, df['Tokyo24/7 R@5'], 's-', label='R@5')
    axes[0].plot(hbm_sizes, df['Tokyo24/7 R@10'], '^-', label='R@10')
    axes[0].set_title('Tokyo24/7')
    axes[0].legend(frameon=False)
    axes[0].set_xlabel('$\lambda$')
    axes[0].set_ylabel('Recall (\\%)')
    axes[0].set_xticks(hbm_sizes)

    # Nordland
    axes[1].plot(hbm_sizes, df['SVOX Night R@1'], 'o-', label='R@1')
    axes[1].plot(hbm_sizes, df['SVOX Night R@5'], 's-', label='R@5')
    axes[1].plot(hbm_sizes, df['SVOX Night R@10'], '^-', label='R@10')
    axes[1].set_title('SVOX Night')
    axes[1].legend(frameon=False)
    axes[1].set_xlabel('$\lambda$')
    axes[1].set_ylabel('Recall (\\%)')
    axes[1].set_xticks(hbm_sizes)

    # Pitts
    axes[2].plot(hbm_sizes, df['SVOX Rain R@1'], 'o-', label='R@1')
    axes[2].plot(hbm_sizes, df['SVOX Rain R@5'], 's-', label='R@5')
    axes[2].plot(hbm_sizes, df['SVOX Rain R@10'], '^-', label='R@10')
    axes[2].set_title('SVOX Rain')
    axes[2].legend(frameon=False)
    axes[2].set_xlabel('$\lambda$')
    axes[2].set_ylabel('Recall (\\%)')
    axes[2].set_xticks(hbm_sizes)

    # Tokyo24/7
    axes[3].plot(hbm_sizes, df['Amster Time R@1'], 'o-', label='R@1')
    axes[3].plot(hbm_sizes, df['Amster Time R@5'], 's-', label='R@5')
    axes[3].plot(hbm_sizes, df['Amster Time R@10'], '^-', label='R@10')
    axes[3].set_title('Amster Time')
    axes[3].legend(frameon=False)
    axes[3].set_xlabel('$\lambda$')
    axes[3].set_ylabel('Recall (\\%)')
    axes[3].set_xticks(hbm_sizes)

    # 
    plt.tight_layout()
    plt.savefig('images/threshold_effect.png',dpi=600)
    # 
    plt.show()
```
