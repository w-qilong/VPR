```python
from pytorch_metric_learning.losses import CrossBatchMemory
from losses import MetricLoss

loss_function= MetricLoss(loss_name='MultiSimilarityLoss')
cross_batch_memory=CrossBatchMemory(
    loss=loss_function.loss_fn,
    embedding_size=1024,
    miner=loss_function.miner,
    memory_size=1024,
    decay_lambda=0.02
)
```

```python
import torch

data=torch.randn(128,1024)
labels=torch.randint(0,1000,(128,))

for i in range(15):
    loss, indices_tuple, time_weights = cross_batch_memory(data, labels)


```

```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots

with plt.style.context(['science','grid','vibrant']):
    # 时间间隔
    time = np.linspace(0, 10, 1000)  # 从 0 到 20，共有 1000 个点

    # 不同的衰减系数 λ
    lambda_values = [0.01,0.02, 0.05, 0.1, 0.5]

    # 创建画布
    plt.figure()

    # 绘制不同 λ 下的权重曲线
    for lamb in lambda_values:
        weights = np.exp(-lamb * time)  # 指数衰减公式
        plt.plot(time, weights, label=f'$\lambda$ = {lamb}')

    # 添加图例和标签
    plt.xlabel(r'$\Delta t$')
    plt.ylabel('Weight')
    plt.legend(frameon=False)

    plt.tight_layout()
    plt.savefig('images/time_decay.png', dpi=600)
    

    # 显示图形
    plt.show()
```
